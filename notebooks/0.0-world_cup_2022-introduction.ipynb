# Exploration of discrete variables

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
DATA_RAW = os.path.join(os.getcwd(), os.pardir, 'data', 'raw')
```

```python
plt.style.use('ggplot')
```

```python
(
    pd.read_csv(os.path.join(DATA_RAW, 'Fifa_WC_2022_Match_data.csv'), encoding = "cp1252")
    ==  pd.read_csv(os.path.join(DATA_RAW, 'Fifa_WC_2022_Match_data.csv'), encoding = "ISO-8859-1")
    ).sum().sum()
```

```python
matches = pd.read_csv(os.path.join(DATA_RAW, 'Fifa_WC_2022_Match_data.csv'), encoding = "ISO-8859-1")
```

```python
matches.dtypes
```

```python
matches.describe()
```

```python
matches.head(5)
```

```python
matches.day_of_week.value_counts()
```

```python
fig, ax = plt.subplots(figsize=(12, 7))
matches.day_of_week.value_counts().plot.bar()
plt.title('Days when matches were played', size=18)
plt.ylabel('Count', size=14)
plt.xlabel('Days', size=14)
plt.show()
```

```python
matches.hour.value_counts()
```

```python
fig, ax = plt.subplots(figsize=(12, 7))
matches.hour.value_counts().plot.bar()
plt.title('Time when matches were played', size=18)
plt.ylabel('Count', size=14)
plt.xlabel('Hour', size=14)
plt.show()
```

```python
matches.venue.value_counts()
```

```python
fig, ax = plt.subplots(figsize=(12, 7))
matches.venue.value_counts().plot.bar()
plt.title('Stadiums where matches were played', size=18)
plt.ylabel('Count', size=14)
plt.xlabel('Stadium', size=14)
plt.show()
```
